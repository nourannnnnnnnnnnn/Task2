```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Downloads/SampleSuperstore.csv')
df
```

```python
# Analyze Categories and Sub-Categories
category_sales = df.groupby('Category')['Sales'].sum()
category_profit = df.groupby('Category')['Profit'].sum()
```

```python
# Visualize Category Sales
plt.figure(figsize=(8, 4))
category_sales.plot(kind='bar', color='blue')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize Category Profit
plt.figure(figsize=(8, 4))
category_profit.plot(kind='bar', color='green')
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```
